# Data Science Notebook

A starter Jupyter notebook for data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create sample data
np.random.seed(42)
data = pd.DataFrame({
    'x': range(50),
    'y': np.random.randn(50).cumsum()
})
data.head()
```

```python
# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(data['x'], data['y'])
plt.title('Random Walk')
plt.xlabel('Step')
plt.ylabel('Value')
plt.show()
```
